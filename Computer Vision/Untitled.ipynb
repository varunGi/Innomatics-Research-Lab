```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
faceCascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('modiji.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                  # To avoid purple modiji
plt.imshow(img)
plt.axis('off')
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, 'gray')
plt.axis('off')
```

```python
hsv_gray = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_gray, 'hsv')
plt.axis('off')
```

```python
inverted_image=cv2.bitwise_not(img_gray)
plt.imshow(inverted_image)
```

```python
blurred=cv2.GaussianBlur(inverted_image,(111,111),0)
plt.imshow(blurred)
```

```python
sketch_image=cv2.divide(img_gray,inverted_image,scale=256.0)
plt.imshow(sketch_image)
```

```python
pencil_sketch=cv2.cvtColor(sketch_image,cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch)
plt.axis('off')
plt.show()
```
