```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
# Reading an image
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
koala = cv2.imread('koala.jpg')
plt.imshow(koala)
plt.axis('off')
```

```python
# Converting image to grayscale
panda_gray = cv2.cvtColor(panda, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_gray, 'gray')
plt.axis('off')
```

```python
koala_gray = cv2.cvtColor(koala, cv2.COLOR_BGR2GRAY)
plt.imshow(koala_gray, 'gray')
plt.axis('off')
```

```python
# Converting to RGB (opencv default is BGR)
panda_orig = cv2.cvtColor(panda, cv2.COLOR_BGR2RGB)
plt.imshow(panda_orig)
plt.axis('off')
```

```python
koala_orig = cv2.cvtColor(koala, cv2.COLOR_BGR2RGB)
plt.imshow(koala_orig)
plt.axis('off')
```

```python
# Gaussian Blur
panda2 = cv2.GaussianBlur(panda_gray, (5,5), 1)
plt.imshow(panda2, 'gray')
plt.axis('off')
```

```python
# Highlighting edges
panda3 = cv2.Canny(panda, 150,200)
plt.imshow(panda3, 'gray')
plt.axis('off')
```

```python
panda4 = cv2.dilate(panda3, kernel = (3,3), iterations = 3)
plt.imshow(panda4, 'gray')
plt.axis('off')
```

```python
panda5 = cv2.erode(panda4, kernel = (3,3), iterations = 5)
plt.imshow(panda5, 'gray')
plt.axis('off')
```
