# Bar graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Id=np.arange(1,10)
score=np.arange(20,110,10)
plt.bar(Id,score)
plt.xlabel("Student ID")
plt.ylabel("Score")
plt.show()
```

```python
Id=np.arange(1,10)
score=np.arange(20,110,10)
plt.bar(Id,score,color='r')
plt.xlabel("Student ID")
plt.ylabel("Score")
plt.title("Example Bar Graph")
plt.grid(True)
plt.show()
```

```python
#plotting multiple sets of Data
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,29,40]
plt.bar(x1,y1,label='First',color='g')
plt.bar(x2,y2,label='Second',color='y')
plt.xlabel('Data1')
plt.ylabel('Data2')
plt.title("Bar graph")
plt.legend()
plt.show()
```

```python
#horizontal bar graph
```

```python
Name=['satya','Varun','Vidya','Prabha','Swetha']
age=[29,26,25,24,22]
plt.barh(Name,age,color='yellowgreen')
plt.show()
```

```python
#changing width of a bar
```

```python
plt.figure(figsize=(8,6))#(col,row)
num1=np.array([1,3,5,7,9])
num2=np.array([2,4,6,8,10])
plt.bar(num1,num1**2,width=0.4,color='r')
plt.bar(num2,num2**2,width=0.3,color='c')
plt.grid(True)
plt.show()
```

```python
#stacked vertical bar plots
```

```python
x=['A','B','C','D']
y1=[10,20,10,30]
y2=[20,25,15,25]
#plot the bars in stack manner
plt.bar(x,y1,color='y',label="First Round ")
plt.bar(x,y2,bottom=y1,color='g',label='Second Round')
plt.title("Game")
plt.legend()
plt.show()
```

```python
x=['A','B','C','D']
y1=np.array([10,20,10,30])
y2=np.array([20,25,15,25])
y3=np.array([25,15,20,23])
#plot the bars in stack manner
plt.bar(x,y1,color='r',label="First Round ")
plt.bar(x,y2,bottom=y1,color='b',label='Second Round')#botom=y1
plt.bar(x,y3,bottom=y1+y2,color='g',label='Third Round')#bottom =y1+y2
plt.title("Game")
plt.legend()
plt.show()
```

```python
x=np.arange(5)
y1=[34,56,12,89,67]
y2=[12,56,78,45,90]
y3=[14,23,45,25,89]
width=0.2
plt.bar(x-0.2,y1,width,color='c')
plt.bar(x,y2,width,color='orange')
plt.bar(x+0.2,y3,width,color='green')
plt.xticks(x,['Team A','Team B','Team C','Team D','Team E'])
plt.xlabel('Teams')
plt.ylabel('Scores')
plt.legend(['Round1','Round2','Round3'])
plt.show()
```
