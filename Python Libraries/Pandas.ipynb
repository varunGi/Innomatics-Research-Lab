# Merge Data frames

```python
df1=pd.DataFrame({"ID":['1','2','3','4','5'],'Name':['Satya','Abhi','Prabha','Swathi','Ishika']})
print(df1)
```

```python
df2=pd.DataFrame({"ID":['1','2','6','7','8'],"Score":[100,90,89,78,99]})
df2
```

```python
#inner join
pd.merge(df1,df2,on="ID",how='inner')
```

```python
#left join
pd.merge(df1,df2,on="ID",how="left")
```

```python
#right join
pd.merge(df1,df2,on="ID",how="right")
```

```python
#outer join
pd.merge(df1,df2,on="ID",how="outer")
```
