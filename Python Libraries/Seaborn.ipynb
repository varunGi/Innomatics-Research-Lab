```python
#high end data visualisation library
```

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
```

```python
data
```

```python
#draw line plot 
sns.lineplot(data.sepal_length,data.sepal_width)
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
plt.xlim(5)
plt.ylim(3)
sns.set_style('dark')
plt.show()
```

```python
plt.subplot(2,1,1)
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
sns.set_style('dark')
plt.show()
plt.subplot(2,1,2)
sns.lineplot(x='petal_length',y='petal_width',data=data)
sns.set_style('white')
plt.show()
```

# Color palette

```python
palette=sns.color_palette()
sns.palplot(palette)
plt.show()
```

```python
p=sns.color_palette('Reds',10)
sns.palplot(p)
plt.show()
```

Relatinal Plots

```python
#relpot()
#these are used for visualizing the statistical relationship between the Data points
```

```python
data
```

```python
sns.relplot(x='sepal_width',y='species',data=data)
plt.show()
```

```python
sns.relplot(x='sepal_width',y='petal_width',data=data)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=data)
plt.show()
```

```python
#categorical plots
```

```python
#bar plot
```

```python
sns.barplot(x='species',y='sepal_length',data=data)
plt.show()
```

```python
#countplot()---------counts the categories and returns a count of their occurances
```

```python
sns.countplot(x='species',data=data)
plt.show()
```

```python
#box
```

```python
sns.boxplot(x='sepal_width',y='species',data=data)
plt.show()
```

```python
#distplot
```

```python
sns.distplot(data['sepal_width'])
plt.show()
```

```python
#pairplots
```

```python
#represents pairswise relation acrosss the entire datta frame and supports an additional argument called hue
#for categorical separation
```
